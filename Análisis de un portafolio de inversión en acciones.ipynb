```python
# Definimos los tickers de las acciones y el benchmark, así como las fechas de inicio y fin para la descarga de datos.
# I define the tickers of the stocks and the benchmark, as well as the start and end dates for data download.

# Elegimos las siguientes acciones: 
# I choose the following stocks:

# Apple (AAPL), Microsoft (MSFT), Amazon (AMZN), JPMorgan Chase (JPM) y el fondo indexado Vanguard S&P 500 ETF (VOO) como benchmark.
# Apple (AAPL), Microsoft (MSFT), Amazon (AMZN), JPMorgan Chase (JPM) and the Vanguard S&P 500 ETF (VOO) as a benchmark.

TICKERS = ["AAPL", "MSFT", "AMZN", "JPM", "VOO"] 

# De igual forma definimos el rango de fechas desde el 1 de enero de 2016 hasta el 1 de enero de 2026 para obtener datos históricos.
# Similarly, we define the date range from January 1, 2016 to January 1, 2026 to obtain historical data.

START_DATE = "2016-01-01"
END_DATE = "2026-01-01"

# Construmos una ruta para guardar los datos descargados en la carpeta "data/raw" y creamos la carpeta si no existe.
# I construct a path to save the downloaded data in the "data/raw" folder and create the folder if it does not exist.

RAW_DIR = os.path.join("data", "raw")
os.makedirs(RAW_DIR, exist_ok=True)

# Definimos la función download_prices() para descargar los precios de las acciones y el benchmark;
# desde Yahoo Finance utilizando la librería yfinance. 

# La función toma como parámetros los tickers, la fecha de inicio y la fecha de fin, y devuelve un DataFrame con los precios diarios;
# en formato largo (long format).

# I define the download_prices() function to download the prices of stocks and the benchmark;
# from Yahoo Finance using the yfinance library.   

# The function takes as parameters the tickers, the start date and the end date, and returns a DataFrame with daily prices;
# in long format.

def download_prices(tickers, start, end):
    data = yf.download(
        tickers=tickers,
        start=start,
        end=end,
        auto_adjust=False,
        progress=False,
        group_by="ticker"
    )
    
    # Convertir a formato largo: date, ticker, open, high, low, close, adj_close, volume.
    # Convert to long format: date, ticker, open, high, low, close, adj_close, volume.
    records = []
    for ticker in tickers:
        df_t = data[ticker].copy()
        df_t["ticker"] = ticker
        df_t.reset_index(inplace=True)
        df_t.rename(columns={
            "Date": "date",
            "Open": "open",
            "High": "high",
            "Low": "low",
            "Close": "close",
            "Adj Close": "adj_close",
            "Volume": "volume"
        }, inplace=True)
        records.append(df_t[["date", "ticker", "open", "high", "low", "close", "adj_close", "volume"]])
    
    df_all = pd.concat(records, ignore_index=True)
    return df_all


# Definimos la función main() para ejecutar el flujo principal del script. 
# Esta función descarga los precios de las acciones y el benchmark, guarda los datos en un archivo CSV;
# y muestra un mensaje indicando la ubicación del archivo guardado.
# Muestra las primeras filas del DataFrame.

# I define the main() function to execute the main flow of the script.
# This function downloads the prices of stocks and the benchmark, saves the data in a CSV file;
# and displays a message indicating the location of the saved file.
# It shows the first rows of the DataFrame.

def main():
    df_prices = download_prices(TICKERS, START_DATE, END_DATE)
    output_path = os.path.join(RAW_DIR, "prices_daily.csv")
    df_prices.to_csv(output_path, index=False)
    print(f"Archivo guardado en: {output_path}")
    print(df_prices.head())

if __name__ == "__main__":
    main()
```

##### 5.1 Calculamos retorno diario y retorno acumulado por activo.
##### 5.1 Comulative daily return and total comulative return by asset.

##### 5.1.1 Calculamos el retorno diario por activo.
##### 5.1.1 Comulative daily return by asset.

```python
# Ordenamos el DataFrame por ticker y fecha para asegurarnos de que los datos estén en el orden correcto antes de calcular los retornos diarios.
# Sino, los retornos diarios podrían calcularse incorrectamente si los datos no están ordenados cronológicamente.

# We sort the DataFrame by ticker and date to ensure that the data is in the correct order before calculating daily returns.
# Otherwise, daily returns could be calculated incorrectly if the data is not sorted chronologically.

df_data=df_data.sort_values(['ticker','date']).copy()

# Calculamos retornos diarios por ticker agrupando por ticker, despues utilizamos la función pct_change() para calcular el cambio porcentual;
# diario en el precio ajustado de cierre (adj_close).
# Utilizamos el precio ajustado de cierre (adj_close) para calcular los retornos diarios, ya que este precio tiene en cuenta los dividendos y;
# los splits de acciones, proporcionando una medida más precisa del rendimiento de la acción a lo largo del tiempo.

# We calculate daily returns by ticker by grouping by ticker, then we use the pct_change() function to calculate the daily percentage change;
# in the adjusted closing price (adj_close).
# We use the adjusted closing price (adj_close) to calculate daily returns, as this price takes into account dividends and stock splits,
# providing a more accurate measure of the stock's performance over time.


df_data['retorno_diario'] = df_data.groupby('ticker')['adj_close'].pct_change().copy()
print(df_data['retorno_diario'].shape)

# Rellenamos los valores nulos en la columna 'retorno_diario' con 0 utilizando el método fillna(). 
# Esto es útil para evitar problemas en cálculos posteriores ya que los retornos diarios no pueden ser calculados para el primer día de cada ticker.

# We fill the null values in the 'retorno_diario' column with 0 using the fillna() method.
# This is useful to avoid problems in later calculations since daily returns cannot be calculated for the first day of each ticker.

df_data['retorno_diario']=df_data['retorno_diario'].fillna(0)

# Comprobamos si hay valores nulos en el DataFrame después de rellenar los valores nulos en la columna 'retorno_diario'.
# We check for null values in the DataFrame after filling the null values in the 'retorno_diario' column.

print(df_data.isnull().sum())
```

##### 5.1.2 Calculamos el retorno acumulado por activo.
##### 5.1.2 Cumulative return computation for each asset.

```python
# Calculamos el rendimiento acumulado por activo utilizando la función cumprod() para calcular el producto acumulativo de (1 + retorno_diario);
# y restando 1 para obtener el rendimiento acumulado.

# We calculated the cumulative return for each asset independently, which allows us to analyze the performance of each stock over time.
# We use the cumprod() function to calculate the cumulative product of (1 + daily return) and subtract 1 to obtain the cumulative return.

# Notamos que el rendimiento acumulado se calcula de manera independiente para cada ticker, lo que nos permite analizar el desempeño de cada acción;
#  a lo largo del tiempo.

# We calculate the cumulative return independently for each ticker, which allows us to analyze the performance of each stock over time.

df_data["rendimiento_acumulado"] = (df_data.groupby("ticker")["retorno_diario"].transform(lambda x: (1 + x).cumprod() - 1))
rendimiento_acumulado = df_data[["ticker", "date", "rendimiento_acumulado"]].copy()

# Creamos una nueva columna 'quarter' en el DataFrame rendimiento_acumulado utilizando la función dt.to_period('Q') para convertir la columna;
# 'date' a un período trimestral, con la finalidad de analizar el rendimiento acumulado por trimestre, lo que nos permite observar cómo ha;
# evolucionado el desempeño de cada acción a lo largo del tiempo en intervalos trimestrales.

# I make a new column 'quarter' in the rendimiento_acumulado DataFrame using the dt.to_period('Q') function to convert the 'date' column to a;
# quarterly period, that way we can analyze the cumulative return by quarter, which allows us to see how the performance of each stock has;
# evolved over time in quarterly intervals.

rendimiento_acumulado['quarter'] = rendimiento_acumulado['date'].dt.to_period('Q')
rendimiento_acumulado["quarter"] = rendimiento_acumulado["quarter"].astype(str)


# We sort the DataFrame by ticker and quarter and drop the original 'date' column
# to keep only the relevant fields for quarterly cumulative return analysis.

rendimiento_acumulado=rendimiento_acumulado.sort_values(["ticker", "quarter"],ascending=True).copy()
rendimiento_acumulado=rendimiento_acumulado.drop('date', axis=1)

# Creamos un nuevo DataFrame 'tabla_final' que contiene el último rendimiento acumulado para cada ticker;
# agrupando el DataFrame 'rendimiento_acumulado' por 'ticker' y seleccionando el último valor de 'rendimiento_acumulado'.

# We create a new DataFrame 'tabla_final' that contains the last cumulative return for each ticker;
# by grouping the 'rendimiento_acumulado' DataFrame by 'ticker' and selecting the last value of 'rendimiento_acumulado'.

print(f"Últimos rendimientos acumulados por activo:")
print(f'Last cumulative returns by asset:')
print()
tabla_final = (rendimiento_acumulado.groupby("ticker")["rendimiento_acumulado"].last().sort_values(ascending=False).reset_index())
tabla_porcentaje = tabla_final.copy()
tabla_porcentaje["rendimiento_acumulado"] = tabla_porcentaje["rendimiento_acumulado"] * 100
tabla_porcentaje["rendimiento_acumulado"] = tabla_porcentaje["rendimiento_acumulado"].round(2)
tabla_porcentaje = tabla_porcentaje.rename(columns={"rendimiento_acumulado": "rendimiento_acumulado_pct"})
print(tabla_porcentaje)

# Graficamos el rendimiento acumulado por activo utilizando la librería seaborn para crear un gráfico de líneas (lineplot);
# donde el eje x representa los trimestres, el eje y representa el rendimiento acumulado, y cada línea representa un ticker diferente. 
# Se desactiva la barra de error (errorbar=None) para que el gráfico sea más limpio y fácil de interpretar.

# We plot the cumulative return by asset using the seaborn library to create a line plot;
# where the x-axis represents the quarters, the y-axis represents the cumulative return, and each line represents a different ticker;
#  The error bar (errorbar=None) is disabled to make the plot cleaner and easier to interpret.
print()
plt.figure(figsize=(11, 5))
sns.lineplot(
    data=rendimiento_acumulado,
    x="quarter",
    y="rendimiento_acumulado",
    hue="ticker",
    errorbar=None
)

plt.title("Rendimiento Acumulado por Activo")
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)  
plt.xlabel("Fecha")
plt.xticks(rotation=45)
plt.ylabel("Rendimiento Acumulado")
plt.grid(False)
plt.tight_layout()
plt.legend(
    frameon=False,     # sin cuadro
    borderpad=0,       # sin padding interno
    labelspacing=0.1,  # espacio entre etiquetas
    handlelength=1,    # tamaño de líneas en la leyenda
    loc="upper left",
    bbox_to_anchor=(0,1)
)
plt.show()
```

```python
# Unimos la tabla de retornos diarios con la tabla del portafolio utilizando la función merge() de pandas;
# especificando la columna 'ticker' como clave de unión y el tipo de unión como 'inner'.
# Con la finalidad de calcular el retorno del portafolio, ya que necesitamos combinar los retornos diarios de cada activo;
# con sus respectivos pesos en el portafolio.

# We calculate the portfolio return by merging the daily returns table with the portfolio table using the merge() function from pandas;
# specifying the 'ticker' column as the join key and the join type as 'inner'. That way we can combine the daily returns of each asset;
# with their respective weights in the portfolio to calculate the portfolio return.

df_merge_portfolio = df_data.merge(portfolio, on='ticker',how='inner').reset_index(drop=True).copy()

# Creamo una columan que la que calculamos el retorno del portafolio, multiplicando el retorno diario de cada activo por su peso en el portafolio.
# We create a column in which we calculate the portfolio return by multiplying the daily return of each asset by its weight in the portfolio.

df_merge_portfolio['retorno_portafolio'] = df_merge_portfolio['retorno_diario'] * df_merge_portfolio['weight']

# Mostramos las primeras filas del DataFrame resultante y verificamos si hay valores nulos en el DataFrame después de la unión.
# We display the first rows of the resulting DataFrame and check for null values in the DataFrame after the merge.
print(df_merge_portfolio.head())
print(df_merge_portfolio.isnull().sum())

# Agrupamos por fecha y sumamos los retornos diarios de las acciones para obtener un solo retorno diario del portafolio por día.
# We group by date and sum the daily returns of the stocks to obtain a single daily return for the portfolio per day.

df_portfolio=df_merge_portfolio.groupby('date')['retorno_portafolio'].sum().reset_index().copy()

# Mostramos las primeras filas del DataFrame resultante y verificamos si hay valores nulos en el DataFrame después de la agrupación.

print(df_portfolio.head(10))
print(df_portfolio.isnull().sum())

```

##### 5.4 Sharpe ratio anualizada por activo y por portafolio.
##### 5.4 Annualized Sharpe ratio per asset and the portfolio.

##### 5.4.1 Tasa libre de riesgo (diaria).
##### 5.4.1 Risk‑free rate (daily).

```python
# Para calcular el Sharpe Ratio necesitamos la tasa libre de riesgo.
# En este caso, descargamos la tasa de los T-Bills a 3 meses desde FRED.

# To calculate the Sharpe Ratio, we need the risk‑free rate.
# Here, we download the 3‑month T‑Bill rate from FRED.

load_dotenv()
api_key = os.getenv("FRED_API_KEY")
fred_api_key = api_key
fred = Fred(api_key=fred_api_key)

# Descargar la serie histórica del T-Bill a 3 meses (DGS3MO).
# Download the historical series for the 3‑month T‑Bill (DGS3MO).
rf = fred.get_series('DGS3MO')
rf = rf.to_frame(name='rf')
rf.index = pd.to_datetime(rf.index)

# Convertir la tasa de porcentaje (%) a decimal.
# Convert the rate from percentage (%) to decimal.
rf['rf'] = rf['rf'] / 100

# Alineamos la tasa libre de riesgo con los retornos diarios.
# We aligning the risk‑free rate with the portfolio’s daily returns.

# Reenombramos la columa para evitar conflictos al realizar el merge.
# We renaming the column to avoid conflicts during the merge.

rf = rf.rename(columns={'rf': 'rf_annual'})
df_portfolio = df_portfolio.merge(rf, left_on='date',right_index=True,how='left')

# Copio la tasa anual a una columna más directa
# Moving the annual risk‑free rate into a cleaner column
df_portfolio['rf'] = df_portfolio['rf_annual']

# Elimino la columna temporal
# Dropping the temporary annual column
df_portfolio.drop(columns=['rf_annual'], inplace=True)

# Relleno hacia adelante los días sin tasa (feriados)
# Forward‑filling missing risk‑free values for holidays or non‑trading days
df_portfolio['rf'] =  df_portfolio['rf'].ffill() 

# Convertimos la tasa libre de riesgo anual a diaria, a tasa de FRED es anual, así que la convierto usando 252 días de trading
# We Converted the annual risk‑free rate into a daily rate, FRED provides an annualized rate, so I scale it down using 252 trading days

df_portfolio['rf_daily'] = df_portfolio['rf'] / 252

# Confirmamos que no hay valores nulos utilizando el metodo, .isnull y .sum
# We look for null values, using methos .isnull and .sum()

print(df_portfolio[['date','retorno_portafolio','retorno_acumulado','rf']].isnull().sum())
```

##### 5.4 Sharpe ratio anualizada por activo.
##### 5.4 Annualized Sharpe ratio per asset.

```python
# Unimos los retornos diarios con la tasa libre de riesgo diaria.
# We merge the daily returns with the daily risk‑free rate.

df_sharpe = df_data.merge(df_portfolio[['date', 'rf_daily']],on='date',how='left').copy()

# Rellenamos los valores faltantes de rf_daily para cubrir los días feriados en los que no se publica la tasa.
# We fill missing rf_daily values to account for holidays when the risk‑free rate is not reported.

df_sharpe['rf_daily'] =  df_sharpe['rf_daily'].ffill() 

# Después calculamos el exceso de retorno diario, restando la tasa libre de riesgo diaria al retorno diario.
# Then we compute the daily excess return by subtracting the daily risk‑free rate from the daily return.

df_sharpe['exceso_retorno'] = df_sharpe['retorno_diario'] - df_sharpe['rf_daily']

# Luego calculamos el retorno anualizado y la volatilidad anualizada por activo.
# Next, we annualize both the mean excess return and the volatility for each asset.

ret_anual_activo = df_sharpe.groupby('ticker')['exceso_retorno'].mean() * 252

vol_anual_activo = df_sharpe.groupby('ticker')['exceso_retorno'].std() * (252 ** 0.5)

# Finalmente, determinamos el Sharpe ratio anualizado por activo, dividiendo el retorno anualizado entre la volatilidad anualizada.
# Finally, we compute the annualized Sharpe ratio per asset by dividing annualized return by annualized volatility.

sharpe_activo = ret_anual_activo / vol_anual_activo

# Transformamos los datos para una mejor presentación: cambiamos los nombres de las columnas y redondeamos los valores a dos decimales.
# We transform the data for better presentation: we rename the columns and round the values to two decimals.
sharpe_activo=sharpe_activo.reset_index()
sharpe_activo=sharpe_activo.rename(columns={'ticker':'ticker','exceso_retorno':'sharpe_ratio'}).copy()
sharpe_activo=sharpe_activo.sort_values(ascending=False, by='sharpe_ratio')
sharpe_activo['sharpe_ratio']=sharpe_activo['sharpe_ratio'].round(2)
print(sharpe_activo)

# Graficamos el Sharpe Ratio para todos los activos.
# We plot the Sharpe Ratio for all the assets.

palette = {"VOO": "#0A4C8A", "AMZN": "#0A4C8A", "JPM": "#0A4C8A", "AAPL": "#FFF021", "MSFT":"#FFF021","Portfolio":"Orange"}

plt.figure(figsize=(9,5))
sns.barplot(
    data=sharpe_activo,
    x="ticker",
    y="sharpe_ratio",
    hue='ticker',
    palette=palette,
    legend=False
)

plt.title("Annualized Sharpe Ratio", fontsize=14)
plt.ylabel("Sharpe (Unidades)")
plt.xlabel("")
plt.show()
```

Observamos que las acciones de Microsoft y Apple son las que entregan el mayor rendimiento ajustado por riesgo. Ambas generan aproximadamente 0.90 unidades de retorno adicional por cada unidad de volatilidad, es decir, por cada unidad de riesgo asumido sobre la tasa libre de riesgo. Sin embargo, nuestro benchmark (VOO) no se encuentra muy alejado: ofrece alrededor de 0.70 unidades de retorno adicional por cada unidad de riesgo, lo que indica que también presenta una eficiencia riesgo‑retorno razonablemente buena.

We observe that Microsoft and Apple are the assets that deliver the highest risk‑adjusted returns.
Both generate approximately 0.90 units of additional return for every unit of volatility, meaning for every unit of risk taken above the risk‑free rate. However, our benchmark (VOO) is not far behind. It provides a bit more than 0.70 units of excess return per unit of risk, indicating that it also offers a reasonably strong risk‑return efficiency.

##### 5.4 Sharpe ratio anualizada del portafolio.
##### 5.4 Annualized Sharpe Ratio of the Portfolio.

```python
# Calculamos el exceso de retorno diario sobre la tasa libre de riesgo.
# We compute the daily excess return over the risk-free rate.

exceso_retorno = df_portfolio['retorno_portafolio'] - df_portfolio['rf'].ffill().fillna(0) / 252

# Calculamos el exceso de retorno anualizado del portafolio.
# We calculate the annualized excess return of the portfolio.

exceso_retorno_anualizado = exceso_retorno.mean() * 252

# Calculamos la volatilidad anualizada del portafolio.
# We compute the annualized volatility of the portfolio.

volatilidad_anualizada_portafolio = df_portfolio['retorno_portafolio'].std() * (252**0.5)

# Obtenemos el Sharpe Ratio anualizado del portafolio.
# We obtain the annualized Sharpe Ratio of the portfolio.

sharpe_ratio = exceso_retorno_anualizado / volatilidad_anualizada_portafolio
print(f"Sharpe Ratio del portafolio: {sharpe_ratio:.3f}")
print(f"Portfolio’s Annualized Sharpe Ratio: {sharpe_ratio:.3f}")
print()

# Creamos una DataFrame que contenga el valor del Sharpe Ratio de los activos y del portafolio.
# We create a DataFrame that contains the Sharpe Ratio values of the individual assets and of the overall portfolio.

sharpe_total = pd.DataFrame({"ticker": ["VOO", "AMZN", "JPM", "AAPL", "MSFT","Portfolio"],"value": [0.74, 0.70, 0.74, 0.91, 0.92, sharpe_ratio]})

# Organizamos y mostramos los valores.
# We organize our values.

sharpe_total=sharpe_total.sort_values("value", ascending=False).copy()
sharpe_total=sharpe_total.rename(columns={'ticker':'ticker','value':'sharpe_ratio'}).copy()
sharpe_total=sharpe_total.sort_values(ascending=False, by='sharpe_ratio')
sharpe_total["sharpe_ratio"] = sharpe_total["sharpe_ratio"].round(2)
print(sharpe_total)


# Graficamos el Sharpe Ratio anualizada del portafolio y los activos utilizando un gráfico de barras con la librería seaborn lo que nos permite;
# visualizar cada cuantas unidades obtenemos por unidad de reisgo obtenida.       
# We use a bar chart to visualize the annualized Sharpe RAtio of the portfolio and the Assets, which allows us to see the overall risk;
# level of the portfolio.

palette = {"VOO": "#0A4C8A", "AMZN": "#0A4C8A", "JPM": "#0A4C8A", "AAPL": "#0A4C8A", "MSFT":"#0A4C8A","Portfolio":"Orange"}

plt.figure(figsize=(11,5))
sns.barplot(
    data=sharpe_total,
    x="ticker",
    y="sharpe_ratio",
    hue="ticker",
    palette=palette,
    legend=False
)

plt.title("Annualized Sharpe Ratio", fontsize=14)
plt.ylabel("Sharpe Ratio (Dollars))")
plt.xlabel("")
plt.show()
```

Nuestro portafolio entrega casi una unidad completa de retorno adicional por encima de la tasa libre de riesgo de manera anualizada, lo cual indica una muy buena eficiencia riesgo‑retorno y lo vuelve atractivo desde una perspectiva de inversión.

Our portfolio delivers nearly one full unit of excess return above the risk‑free rate on an annualized basis, which reflects strong risk‑adjusted performance and makes it an attractive investment from a risk‑return efficiency standpoint.

##### 5.5 Máximo drawdown por activo y por portafolio.  
##### 5.5 Maximum Drawdown per Asset and for the Portfolio.

##### 5.5.1 Máximo drawdown por activo.  
##### 5.5.1 Maximum Drawdown per Asset.

```python
# Calculamos el valor acumulado (cumulative return) para cada activo.
# We compute the cumulative return for each asset.

df_data['valor'] = df_data.groupby('ticker')['retorno_diario'].transform(lambda x: (1 + x).cumprod())

# Calculamos el máximo histórico (peak) alcanzado por cada activo.
# We calculate the historical peak reached by each asset.

df_data['peak'] = df_data.groupby('ticker')['valor'].transform('cummax')

# Calculamos el drawdown como la caída porcentual desde el máximo histórico.
# We compute drawdown as the percentage decline from the historical peak.

df_data['drawdown'] = df_data['valor'] / df_data['peak'] - 1

# Calculamos el máximo drawdown por activo.
# We calculate the maximum drawdown for each asset.

max_dd_activos = df_data.groupby('ticker')['drawdown'].min().reset_index().copy()
max_dd_activos=max_dd_activos.sort_values(ascending=True,by="drawdown")
max_dd_activos['drawdown']=max_dd_activos['drawdown']*100
max_dd_activos['drawdown']=max_dd_activos['drawdown'].round(2)
print()
print(max_dd_activos)
print()

# Graficamos Max Drawdown para todos los activos.
# We plot the Max Drawdown for all the assets.

palette = {"VOO": "#0A4C8A", "AMZN": "#FFF021", "JPM": "#0A4C8A", "AAPL": "#0A4C8A", "MSFT":"#0A4C8A","Portfolio":"Orange"}

plt.figure(figsize=(9,5))
sns.barplot(
    data=max_dd_activos,
    x="ticker",
    y="drawdown",
    hue='ticker',
    palette=palette,
    legend=False
)

plt.title("Max Drawdown", fontsize=14)
plt.ylabel("Max Drawdown (%)")
plt.xlabel("")
plt.show()
```

Amazon shows the deepest drawdown, with a maximum decline of 56%, compared to the benchmark’s 33%. This indicates that Amazon has experienced a larger relative loss from its historical peak than the reference index.

#### 6. Pruebas Estadisticas.

##### 6.1 Prueba de Levene para comparar varianzas (volatilidad)
##### 6.1 Levene test to compare variances (volatility)

```python
# Retornos del portafolio y del benchmark
# Portfolio and benchmark returns
ret_port = df_valores['ret_portafolio'].dropna()
ret_voo = df_ret['VOO'].dropna()

# Prueba de Levene: compara si las varianzas son iguales
# Levene test: checks if variances are equal
stat_levene, p_levene = levene(ret_port, ret_voo)

print(f"Estadístico de Levene: {stat_levene:.4f}")
print(f"p-value: {p_levene:.4f}")

# Interpretación de resultados
# Interpretation of results
if p_levene < 0.05:
    print("Las varianzas son significativamente diferentes (volatilidad distinta). / The variances are significantly different (different volatility).")
else:
    print("No hay evidencia estadística de diferencia en las varianzas (volatilidad similar). / There is no statistical evidence of a difference in variances (similar volatility).")
```

##### 6.2 Prueba de Diferencia de Medias (t‑test).
##### 6.2 Statistical Tests and Mean Difference Test (t‑test). 

```python
# Retornos del portafolio y del benchmark
# Portfolio and benchmark returns (cleaning NaN values)

ret_port = df_valores['ret_portafolio'].dropna()
ret_voo = df_ret['VOO'].dropna()

# Prueba t de diferencia de medias
# t-test for difference in means between portfolio and benchmark returns

t_stat, p_value = ttest_ind(ret_port, ret_voo, equal_var=False)

print(f"Estadístico t: {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")

# Interpretación
# Interpretation of statistical significance
if p_value < 0.05:
    print("La diferencia de medias es estadísticamente significativa. / The difference in means is statistically significant.")
else:
    print("No hay evidencia estadística de diferencia en las medias. / There is no statistical evidence of a difference in means.")
```
